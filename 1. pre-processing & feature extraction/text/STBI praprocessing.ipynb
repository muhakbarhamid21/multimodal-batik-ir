```python
import pandas as pd
import nltk
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
stop_words = set(stopwords.words('indonesian'))
```

```python
def preprocess_text(text):
    if pd.isna(text):
        return ""

    # Tokenization
    tokens = word_tokenize(text.lower())

    # Stop word removal
    tokens = [word for word in tokens if word not in stop_words]

    # Lemmatization (menggunakan stemmer Sastrawi)
    lemmatized_tokens = [stemmer.stem(word) for word in tokens]

    # Menggabungkan kembali token yang sudah diproses
    return ' '.join(lemmatized_tokens)
```

```python
df = pd.read_csv("/content/data_text_batik.csv")
```

```python
df['processed_nama'] = df['Nama'].apply(preprocess_text)
df['processed_deskripsi'] = df['Deskripsi'].apply(preprocess_text)

# Menyimpan hasil ke file baru
df.to_csv("preprocessed_output.csv", index=False)
```
